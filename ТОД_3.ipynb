2.1 На основе файла contributors_sample.json создайте словарь следующего вида:

{
    должность: [список username людей, занимавших эту должность]
}

```python
import json
with open('contributors_sample.json') as j:
    j = json.load(j)
dic = {}
for person in j:
    for job in person['jobs']:
        if job in dic.keys():
            dic[job] += [person['username']]
        else:
            dic[job] = [person['username']]
for i in dic:
    dic[i] = list(set(dic[i]))
dic
```

3.1 По данным файла steps_sample.xml сформируйте словарь с шагами по каждому рецепту вида {id_рецепта: ["шаг1", "шаг2"]}. Сохраните этот словарь в файл steps_sample.json

```python
from bs4 import BeautifulSoup
with open('steps_sample.xml') as f:
    tree = BeautifulSoup(f, 'xml')
dic = {}
for recipe in tree.recipes.find_all('recipe'):
    st = [steps.next for steps in recipe.steps.find_all("step")]
    dic[int(recipe.find('id').text)] = st
for key, val in dic.items():
    print(f'{key} : {val}')
```

3.2 По данным файла steps_sample.xml сформируйте словарь следующего вида: кол-во_шагов_в_рецепте: [список_id_рецептов]

```python
dic_steps = {}
for el in dic:
    dic_steps[el] = len(dic[el])
dic_steps
dic1 = {}
for key, val in dic_steps.items():
    if val in dic1.keys():
        dic1[val] += [key]
    else:
        dic1[val] = [key]
for key, val in dic1.items():
    print(f'{key} : {val}')
```

3.3 Получите список рецептов, в этапах выполнения которых есть информация о времени (часы или минуты). Для отбора подходящих рецептов обратите внимание на атрибуты соответствующих тэгов.

```python
id_recipe = []
for recipe in tree.find_all('recipe'):
  for steps in recipe.find_all('steps'):
    if 'has_hours' in steps.step.attrs or 'has_minutes' in steps.step.attrs:
      id_recipe.append(recipe.find('id').next)
print(id_recipe)
```
